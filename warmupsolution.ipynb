```python
import pandas as pd 

cars_df = pd.read_csv("/home/student/mycode/clean-with-pandas/cars1.csv")
cars_df2 = pd.read_csv("/home/student/mycode/clean-with-pandas/cars2.csv")
allcars= pd.concat([cars_df, cars_df2], ignore_index=True, sort=False)
allcars.head(2)
```

```python
allcars.set_index("Stock #", inplace=True)
allcars.head(2)
```

```python
allcars[["VIN", "Model", "Price", "timestamp"]]
```

### Find the top 10 most expensive cars

```python
sorted_by_cost = allcars.sort_values(["Price"], ascending=False)
sorted_by_cost.head(10)["Model"]
```

```python
allcars[["VIN", "Model", "Price", "timestamp"]].sort_values("Price", ascending = False).head(10)
```

### Which car has the least amount of miles on it (lowest odometer)

```python
sorted_by_miles = allcars.sort_values(["Odometer"], ascending=True)
sorted_by_miles.head(5)
```

```python
allcars[allcars['Odometer'] == allcars['Odometer'].min()]
```

### If all the cars on the lot were swept away in a tornado, how much money would we lose?

```python
allcars["Price"].sum()
```

```python
print("If all cars were damaged, the total loss is", round(allcars["Price"].sum()), "dollars.")
```
