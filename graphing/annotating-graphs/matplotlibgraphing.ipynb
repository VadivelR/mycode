# Customizing Matplotlib Graphs


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data we want to plot. Using numpy's linspace function, we are creating an array of 70 values from -2pi to 2pi
X = np.linspace(-2 * np.pi, 2 * np.pi, 70, endpoint=True)
F1 = np.sin(2* X)
F2 = (2*X**5 + 4*X**4 - 4.8*X**3 + 1.2*X**2 + X + 1)*np.exp(-X**2)

fig, ax = plt.subplots()

# Using 'none' as the color parameter will make the spines invisible:
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Reposition the bottom spine to y=0:
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

# Reposition left spine to x= 0:
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

ax.plot(X, F1, X, F2)

plt.show()
```

```python
fig, ax = plt.subplots()

# Get the default xticks and labels from the ax object created above
xticks = ax.get_xticks()
xticklabels = ax.get_xticklabels()
print(xticks, xticklabels)

# Print out the first 6 xtick labels
for i in range(6):
    print(xticklabels[i])

yticks = ax.get_yticks()
print(yticks)
```
