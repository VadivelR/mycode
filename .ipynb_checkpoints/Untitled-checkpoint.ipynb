```python
import pandas as pd

cars1 = pd.read_csv("cars1.csv")
cars2 = pd.read_csv("cars2.csv")

cars = pd.concat([cars1,cars2])
cars.shape # to check the total count

cars = cars.set_index('Stock #')
cars

#Find the top 10 most expensive cars
sorted_cars_by_expensice = cars.sort_values(["Price"], ascending=False) # sort by price 
sorted_cars_by_expensice.head(10) #display top 10
```
