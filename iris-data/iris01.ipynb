```python
# python3 -m pip install matplotlib
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
x_index = 3   # this maps to 'petal width (cm)'
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)


ax.set_xlabel(iris.feature_names[x_index])

# where the legend will appear
ax.legend(loc='upper right')

# create the histogram
plt.show()
```

```python
### display sepal width (cm) ###

fig, ax = plt.subplots()

## this is the value that controls the type of data to plot 
x_index = 1   # this maps to 'sepal width (cm)'

colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='upper right')
plt.show()


### display petal length (cm) ###

fig, ax = plt.subplots()

## this is the value that controls the type of data to plot 
x_index = 2   # this maps to 'petal length (cm)'

colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.hist(iris.data[iris.target==label, x_index], 
            label=iris.target_names[label],
            color=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.legend(loc='upper right')
plt.show() 
```

```python
# scatter plot with two features
# sepal length (cm) v petal width (cm)

# python3 -m pip install matplotlib
# import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x_index = 3   # this maps to 'petal width (cm)'
y_index = 0   # this maps to 'sepal length (cm)'

# colors to map on our dots
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.scatter(iris.data[iris.target==label, x_index], 
                iris.data[iris.target==label, y_index],
                label=iris.target_names[label],
                c=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.set_ylabel(iris.feature_names[y_index])
ax.legend(loc='upper left')
plt.show()
```

```python
# scatter plot with two features
# petal length (cm) v sepal width (cm)

# python3 -m pip install matplotlib
# import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x_index = 1   # this maps to 'sepal width (cm)'
y_index = 2   # this maps to 'petal length (cm)'

# colors to map on our dots
colors = ['blue', 'red', 'green']

for label, color in zip(range(len(iris.target_names)), colors):
    ax.scatter(iris.data[iris.target==label, x_index], 
                iris.data[iris.target==label, y_index],
                label=iris.target_names[label],
                c=color)

ax.set_xlabel(iris.feature_names[x_index])
ax.set_ylabel(iris.feature_names[y_index])
ax.legend(loc='upper left')
plt.show()
```

```python
# python3 -m pip install matplotlib
# import matplotlib.pyplot as plt

n = len(iris.feature_names)
fig, ax = plt.subplots(n, n, figsize=(16, 16))

colors = ['blue', 'red', 'green']

for x in range(n):
    for y in range(n):
        xname = iris.feature_names[x]
        yname = iris.feature_names[y]
        for color_ind in range(len(iris.target_names)):
            ax[x, y].scatter(iris.data[iris.target==color_ind, x], 
                             iris.data[iris.target==color_ind, y],
                             label=iris.target_names[color_ind],
                             c=colors[color_ind])

        ax[x, y].set_xlabel(xname)
        ax[x, y].set_ylabel(yname)
        ax[x, y].legend(loc='upper left')


plt.show()
```
